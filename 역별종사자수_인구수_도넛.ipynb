```python
df_workers = pd.read_csv("./data/grid_통계/2022년_종사자_다사_100M.txt", sep = "^", header = None)
```

```python
df_popul = pd.read_csv("./data/grid_통계/2023년_인구_다사_100M.txt", sep = "^", header = None)
```

```python
# 전체 인구수만
df_popul = df_popul[df_popul[2] == "to_in_001"]
df_popul.head()
```

```python
metro_df = gpd.read_file("./data/서울지하철_shp/seoul_subsway_station_4326.shp", encoding = "cp949")
```

```python
# 격자의 중심점 구하기
df["centroid"] = df.geometry.centroid
```

```python
buffer_1000 = metro_df.to_crs(epsg=5179).geometry.buffer(1000)
```

```python
buffer_1000_df = gpd.GeoDataFrame(buffer_1000)
```

```python
buffer_1000_df.head()
```

```python
buffer_1000_df = buffer_1000_df.set_geometry(0)
```

```python
buffer_1000_df.head()
```

```python
buffer_500 = metro_df.to_crs(epsg=5179).geometry.buffer(500)
```

```python
buffer_500_df = gpd.GeoDataFrame(buffer_500)
```

```python
buffer_500_df = buffer_500_df.set_geometry(0)
```

```python
res_difference = gpd.overlay(buffer_1000_df, buffer_500_df, how='difference')
```

```python
res_difference = res_difference[0]
```

```python
for i in range(len(metro_df)):
    print(res_difference[i].contains(df["centroid"]).sum())
```

```python
metro_grid_dict = dict()
for i in range(len(metro_df)):
    metro_grid_lst = df[res_difference[i].contains(df["centroid"])]["GRID_100M_"].values
    metro_grid_dict[metro_df.loc[i, "지하철역"]] = metro_grid_lst
```

```python
metro_grid_dict
```

```python
len(metro_grid_dict)
```

```python
df_workers[df_workers[1].isin(metro_grid_dict["가산디지털단지"])][3].sum()
```

```python
df_popul[df_popul[1].isin(metro_grid_dict["가산디지털단지"])][3].sum()
```

```python
df_workers[df_workers[1].isin(metro_grid_dict["시청"])][3].sum()
```

```python
df_popul[df_popul[1].isin(metro_grid_dict["시청"])][3].sum()
```

```python
worker_dict = dict()
for key in metro_grid_dict:
    worker_dict[key] = df_workers[df_workers[1].isin(metro_grid_dict[key])][3].sum()
```

```python
workers_sr = pd.Series(worker_dict)
```

```python
len(worker_dict)
```

```python
popul_dict = dict()
for key in metro_grid_dict:
    popul_dict[key] = df_popul[df_popul[1].isin(metro_grid_dict[key])][3].sum()
```

```python
popul_sr = pd.Series(popul_dict)
```

```python
len(popul_dict)
```

```python
result = pd.concat([workers_sr, popul_sr], axis = 1)
```

```python
result.columns = ["종사자수", "거주인구수"]
```
